## 2. Configuration Space Exploration

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle, Circle
import numpy as np

# Define the obstacles: rectangles by corners, circles by center and radius
rects = [((2, 2), (3, 4)), ((6, 5), (7, 7))]
circles = [((4, 8), 0.6), ((8, 2), 1.0)]
robot_radius = 0.5

# Inflate rectangular obstacles
def inflate_rectangles(rectangles, r):
    inflated = []
    for (x1, y1), (x2, y2) in rectangles:
        inflated.append(((x1 - r, y1 - r), (x2 + r, y2 + r)))
    return inflated

# Inflate circular obstacles
def inflate_circles(circles, r):
    return [(center, radius + r) for (center, radius) in circles]

# Begin plotting
fig, ax = plt.subplots()

# Plot original rectangles
for i, ((x1, y1), (x2, y2)) in enumerate(rects):
    w, h = x2 - x1, y2 - y1
    ax.add_patch(Rectangle((x1, y1), w, h, color='gray', label='Original Rect' if i == 0 else ""))

# Plot original circles
for i, (center, radius) in enumerate(circles):
    ax.add_patch(Circle(center, radius, color='blue', alpha=0.6, label='Original Circle' if i == 0 else ""))

# Plot inflated rectangles
inflated_rects = inflate_rectangles(rects, robot_radius)
for i, ((x1, y1), (x2, y2)) in enumerate(inflated_rects):
    w, h = x2 - x1, y2 - y1
    ax.add_patch(Rectangle((x1, y1), w, h, color='orange', alpha=0.3, label='Inflated Rect' if i == 0 else ""))

# Plot inflated circles
inflated_circles = inflate_circles(circles, robot_radius)
for i, (center, radius) in enumerate(inflated_circles):
    ax.add_patch(Circle(center, radius, color='red', alpha=0.3, label='Inflated Circle' if i == 0 else ""))

# Set limits to ensure all shapes are visible
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# Final plot settings
ax.set_title("Inflated Obstacles (Configuration Space)")
ax.set_aspect('equal')
ax.grid(True)
ax.legend(loc='upper right')
plt.show()
```
