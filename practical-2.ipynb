## 2. Configuration Space Exploration

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

# Define both rectangular and circular obstacles
rect_obstacles = [((2, 2), (3, 4)), ((6, 5), (7, 7))]
circle_obstacles = [((4, 8), 0.6), ((8, 2), 1.0)]
robot_radius = 0.5

# Inflate rectangular obstacles
def inflate_rectangles(obs, r):
    inflated = []
    for (x1, y1), (x2, y2) in obs:
        inflated.append(((x1 - r, y1 - r), (x2 + r, y2 + r)))
    return inflated

# Inflate circular obstacles
def inflate_circles(obs, r):
    inflated = []
    for (center, rad) in obs:
        inflated.append((center, rad + r))
    return inflated

# Plot
fig, ax = plt.subplots()

# Plot original rectangular obstacles
for (x1, y1), (x2, y2) in rect_obstacles:
    ax.add_patch(patches.Rectangle((x1, y1), x2 - x1, y2 - y1, color='gray', label='Original Rect'))

# Plot original circular obstacles
for (center, radius) in circle_obstacles:
    ax.add_patch(patches.Circle(center, radius, color='blue', alpha=0.6, label='Original Circle'))

# Plot inflated rectangular obstacles
inflated_rects = inflate_rectangles(rect_obstacles, robot_radius)
for (x1, y1), (x2, y2) in inflated_rects:
    ax.add_patch(patches.Rectangle((x1, y1), x2 - x1, y2 - y1, color='orange', alpha=0.3, label='Inflated Rect'))

# Plot inflated circular obstacles
inflated_circles = inflate_circles(circle_obstacles, robot_radius)
for (center, radius) in inflated_circles:
    ax.add_patch(patches.Circle(center, radius, color='red', alpha=0.3, label='Inflated Circle'))

# Plot settings
plt.title("Configuration Space with Inflated Obstacles")
plt.axis('equal')
plt.grid(True)

# Avoid duplicate labels in legend
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys(), loc='upper right')

plt.show()
```
