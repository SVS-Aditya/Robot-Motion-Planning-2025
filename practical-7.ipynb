## 7. SLAM: Simultaneous Localization and Mapping
Simulated SLAM using noisy robot motion and noisy landmark observations.

```python

import matplotlib.pyplot as plt
import numpy as np

landmarks = np.array([[5, 10], [10, 5], [15, 15]])
trajectory = []
est_trajectory = []
pose = np.array([0, 0])
est_pose = np.array([0, 0])
noise = lambda: np.random.randn() * 0.5

for t in range(15):
    u = np.array([1, 1])
    pose = pose + u
    z = landmarks - pose + noise()
    est_pose = est_pose + u + noise()
    trajectory.append(pose.copy())
    est_trajectory.append(est_pose.copy())

trajectory = np.array(trajectory)
est_trajectory = np.array(est_trajectory)

plt.plot(trajectory[:,0], trajectory[:,1], label="True Path")
plt.plot(est_trajectory[:,0], est_trajectory[:,1], label="Estimated Path")
plt.scatter(*landmarks.T, c='red', label="Landmarks")
plt.title("Simulated SLAM")
plt.legend()
plt.grid(True)
plt.show()
```
